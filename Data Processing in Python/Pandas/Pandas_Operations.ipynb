```python
import pandas as pd
```

```python
data = {"Name":['Shaswata','Payel'],'Age':[26,25], "Location":["Bangalore","Kolkata"]}
df = pd.DataFrame(data)
df
```

Selecting and Indexing

```python
df["Name"]  # Retriving Columns
```

```python
df[["Name","Age"]]
```

```python
df.iloc[0] #Retriving Rows
```

```python
df[0:1] # Slicing
```

Filtering

```python
df[df['Age']>25]
```

```python
df[(df['Age']<26) & (df['Location'] == 'Kolkata')]
```

Data Cleaning

```python
df.isnull()
```

```python
df.dropna() #Eleminate Null Values
```

```python
df.fillna(0)  # Fill null values with 0
```

Rename Column

```python
df.rename(columns={'Location':'Current Location'}, inplace=True)
df
```

Replacing Values

```python
df.replace('Shaswata','Shaswata Barua', inplace=True)
df
```

Mathametical Operation

```python
df['Age'].mean()
df['Age'].sum()
df['Age'].max()
df['Age'].min()
#df['Age'].count()
```

Sorting

```python
df.sort_values(by='Age')  # Default is Assending Order
```

```python
df.sort_values(by='Age', ascending=False)
```

Grouping

```python
df.groupby('Name')['Age'].mean()
```
