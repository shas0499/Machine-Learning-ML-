Pandas is a Python library, primarily use for data manupulation and data analysis

```python
import pandas as pd
```

Create Searies

```python
srs = pd.Series([10,20,30,40])
print(srs)

#Edit Index
srs1 = pd.Series([10,20,30,40], index=['a','b','c','d'])
print(srs1)
```

Dataframe

```python
data = {"Name":['Shaswata','Payel'],'Age':[26,25]}
df = pd.DataFrame(data)
df
```

Create dataframe from list of Dictionary

```python
data = [{'Name' : "Shaswata",'Age':26},{'Name' : "Payel",'Age':25}]
df = pd.DataFrame(data)
df
```
