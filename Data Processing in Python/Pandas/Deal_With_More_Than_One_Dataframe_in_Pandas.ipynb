```python
import pandas as pd
```

Marging Two Dataframes

```python
df1 = pd.DataFrame({'EmpID':['EMP001','EMP002','EMP003'],'Name':['Shaswata','Prantika','Sudhir']})
df2 = pd.DataFrame({'EmpID':['EMP001','EMP002','EMP003'], 'Salary':[90000,70000,95000]})
print(df1)
print(df2)
print('After Merging')
marged = pd.merge(df1,df2)
print(marged)
```

Concatination

```python
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=1)
```
