```python
import numpy as np
```

```python
arr =  np.array([[1,2,3],[10,20,30]])
arr
```

Indexing

```python
print(arr[1,2])
print(arr[0,2])
print(arr[0,:])
```

Array Operations

```python
arr1 = np.array([15,25,35])
arr2 = np.array([20,40,50])
print(arr1+arr2)
print(arr1-arr2)
print(arr1*arr2)
print(arr1**2)
print(arr2**2)
```

Comparison Operations

```python
arr3 = np.array([3,10,2,6])
print(arr3>5)
```

Universal Functions

```python
np.sqrt([2,4,9])
np.exp([2,4,9])
np.sin([0,np.pi/2])
```

Broadcasting

one array is 1D and 2nd array is 2D, then elements will 2D array will operaation on 1D array

```python
arr4 = np.array([[1],[2],[3]])
arr5 = np.array([5,6,7])
print(arr5+arr4)
```

Reshaping

```python
arr6 = np.array([[1,2],[3,4],[5,6]])
arr6
print(arr6.reshape(2,3))

#Flattening : All the elements will be in single array
print(arr6.flatten())
```
