Aggregation & Statistics

```python
import numpy as np
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print(np.sum(arr))
print(np.mean(arr))
print(np.max(arr))
print(np.min(arr))
```

Matrix Operations. (Dot Product Operartions)

```python
arr1 = np.array([[1,2],[4,5]])
arr2 = np.array([[1,3],[4,6]])
print(arr1.dot(arr2))
```
