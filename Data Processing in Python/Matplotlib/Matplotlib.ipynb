```python
import matplotlib.pyplot as plt
from matplotlib.lines import lineStyles
```

Basic Line Plot

```python
x = [2,4,6,8,10]
y = [13,17,19,23,29]
plt.plot(x,y)
plt.title('Even - Prime Number')
plt.xlabel('Even Number')
plt.ylabel('Prime Number')
plt.show()
```

Customized Line Plot

```python
x = [2,4,6,8,10]
y = [10,2,30,4,40]
plt.plot(x,y,color = 'Green',linestyle='--',marker='o',linewidth=2)
plt.title('---Customized Line Plot---')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.show()
```

Barchart

```python
student = ['Shaswata','Sudhir','Karan']
marks = [90,95,70]
plt.bar(student,marks,color='skyblue')
plt.title('Student Marks')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.show()
```

SkaterPlot

```python
x = [10,20,30,40,50]
y = [5,15,60,45,100]
plt.scatter(x,y,color='red')
plt.title('Scatter Plot')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.show()
```
