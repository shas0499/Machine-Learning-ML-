Creating an array

```python
arr = np.array([10, 20, 30, 40])
print(arr)
```

Array Dimention

```python
print(arr.ndim)
```

Shape of an Array

```python
print(arr.shape)
```

Size of an Array

```python
print(arr.size)
```

DataType

```python
print(arr.dtype)
```

2-D Array

```python
arr1 = np.array([[10,20],[30,40]])
print(arr1)
```

Operations

```python
print(arr1.ndim)
print(arr1.shape)
print(arr1.size)
print(arr1.dtype)
```
