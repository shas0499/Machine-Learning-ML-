```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = {'Experience':[1,2,3,4,5],'Salary':[25000,45000,60000,75000,90000]}
df = pd.DataFrame(data)
x = df['Experience']  #Feature
y = df['Salary']      #Target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2) #Split the data. 20% data will be in testing data.
```

```python
model = LinearRegression()
model.fit(x_train.to_frame(), y_train)
```

```python
y_pred = model.predict(x_test.to_frame())
print(y_pred)
```

Plot the Real vs Predicted Value

```python
import matplotlib.pyplot as plt
plt.scatter(x_test, y_test,color='blue',label='actual')
plt.scatter(x_test,y_pred,color='red',label='predicted')
#When you use plt.plot(x_test, y_pred), you are telling Matplotlib:
  #"Go to this point on the 'Experience' axis (x_test)"
  #"Go to this point on the 'Salary' axis (y_pred)"
  #"Draw a line connecting these coordinates."
#If you used y_test instead of x_test, you would be trying to plot Salary vs. Salary, which
#would just result in a diagonal line (if the predictions were perfect) and would totally lose the
#context of Experience.
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Actual vs predicted')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(mean_squared_error(y_test,y_pred))
```

Here mean squared is much high so the error is occured. We need to refine our model.
